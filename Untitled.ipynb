```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from matplotlib.patches import Patch
import networkx as nx
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
import time, warnings
warnings.filterwarnings("ignore")

plt.rcParams.update({
    "figure.dpi": 130, "font.size": 10,
    "axes.spines.top": False, "axes.spines.right": False,
    "axes.grid": True, "grid.alpha": 0.25,
})

# Palette couleurs
C = {
    "hw":      "#E74C3C",  # Autoroute
    "nm":      "#3498DB",  # Nationale
    "ur":      "#9B59B6",  # Urbain
    "true":    "#2C3E50",  # Valeur réelle
    "gcn":     "#8E44AD",  # Prédiction GCN
    "err":     "#E67E22",  # Erreur
    "peak":    "#C0392B",  # Heure de pointe
    "offpeak": "#2980B9",  # Heure creuse
    "weekend": "#27AE60",  # Week-end
}
print("Imports OK")
```

```python
# ══════════════════════════════════════════════════════════
# Dataset PEMS-BAY synthétique — 325 capteurs, 6 mois, 5min
# ══════════════════════════════════════════════════════════
np.random.seed(42)
N, NS = 52416, 325
t = np.arange(N)

# Signal de base : cycle journalier + hebdo + heures de pointe
base = np.clip(
    28*np.sin(2*np.pi*t/288 - np.pi/2) + 48
    + 6*np.sin(2*np.pi*t/(288*7))
    - 18*np.exp(-0.5*((t%288 - 96)/9)**2)    # pointe matin
    - 14*np.exp(-0.5*((t%288 - 204)/9)**2)   # pointe soir
    + np.where((t//288)%7 >= 5, 8, 0),        # bonus weekend
    5, 72)

# 3 types de route avec offsets de vitesse
hw = np.random.choice(NS, 100, replace=False)
ur = np.random.choice([i for i in range(NS) if i not in hw], 80, replace=False)
nm = np.array([i for i in range(NS) if i not in hw and i not in ur])

so = np.random.uniform(-5, 5, NS)
so[hw] += 10   # autoroutes plus rapides
so[ur] -= 8    # urbain plus lent

# Corrélation spatiale entre voisins
dv = base[:,None] + so[None,:]
for i in range(1, NS-1):
    dv[:,i] = 0.7*dv[:,i] + 0.15*dv[:,i-1] + 0.15*dv[:,i+1]
dv += np.random.normal(0, 2.0, (N, NS))
dv  = np.clip(dv, 2, 75).astype(np.float32)

didx = pd.date_range("2017-01-01", periods=N, freq="5min")
df   = pd.DataFrame(dv, index=didx, columns=[f"S{i:03d}" for i in range(NS)])

# Split chronologique : Train Jan-Avr / Test Mai-Juin
trdf = df.loc[:"2017-04-30 23:55"]
tedf = df.loc["2017-05-01":]

print(f"TRAIN : {len(trdf):,} pts  {trdf.index[0].date()} → {trdf.index[-1].date()}")
print(f"TEST  : {len(tedf):,} pts  {tedf.index[0].date()} → {tedf.index[-1].date()}")
print(f"Types : {len(hw)} autoroutes | {len(nm)} nationales | {len(ur)} urbains")
```

```python
# ══════════════════════════════════════════════════════════
# Normalisation MinMax par capteur + séquences glissantes
# ══════════════════════════════════════════════════════════
SL = 12   # fenêtre = 1h (12 × 5min)

scalers = []
trs = np.zeros(trdf.shape, np.float32)
tes = np.zeros(tedf.shape, np.float32)
for i in range(NS):
    sc = MinMaxScaler()
    trs[:,i] = sc.fit_transform(trdf.values[:,i:i+1]).flatten()
    tes[:,i] = sc.transform(tedf.values[:,i:i+1]).flatten()
    scalers.append(sc)

# 20 capteurs représentatifs (7 auto + 7 nationale + 6 urbain)
np.random.seed(7)
gcn_s = np.sort(np.concatenate([
    np.random.choice(hw, 7, replace=False),
    np.random.choice(nm, 7, replace=False),
    np.random.choice(ur, 6, replace=False),
]))
NG = len(gcn_s)
tg,  teg      = trs[:, gcn_s], tes[:, gcn_s]
gcn_types     = np.array(["hw" if s in hw else "ur" if s in ur else "nm" for s in gcn_s])
gcn_hw, gcn_nm, gcn_ur = (gcn_types=="hw"), (gcn_types=="nm"), (gcn_types=="ur")

# Séquences glissantes (n_windows, N_capteurs, seq_len)
def make_seq(data, sl=12):
    X, y = [], []
    for i in range(len(data) - sl):
        X.append(data[i:i+sl].T)
        y.append(data[i+sl])
    return np.array(X, np.float32), np.array(y, np.float32)

Xgtr, ygtr = make_seq(tg,  SL)
Xgte, ygte = make_seq(teg, SL)
print(f"Train : {Xgtr.shape}   Test : {Xgte.shape}")
```

```python
# ══════════════════════════════════════════════════════════
# Matrice adjacence par corrélation + normalisation Laplacienne
# ══════════════════════════════════════════════════════════
corr  = np.corrcoef(tg[:5000].T)
A_raw = np.where(corr > 0.65, corr, 0).astype(np.float32)
np.fill_diagonal(A_raw, 0)

# Laplacien symétrique normalisé : D^{-1/2} (A+I) D^{-1/2}
Ah  = A_raw + np.eye(NG, dtype=np.float32)
AN  = (np.diag(Ah.sum(1)**-0.5) @ Ah @ np.diag(Ah.sum(1)**-0.5)).astype(np.float32)
ned = int((A_raw > 0).sum() // 2)
print(f"Graphe : {NG} noeuds, {ned} arêtes (corrélation > 0.65)")
```

```python
# ══════════════════════════════════════════════════════════
# Entraînement GCN NumPy from scratch (~80s)
# Architecture : GraphConv(16,ReLU) → GraphConv(12,ReLU) → Dense(1)
# ══════════════════════════════════════════════════════════
H1, H2 = 16, 12
np.random.seed(99)
W1 = np.random.uniform(-np.sqrt(2/SL), np.sqrt(2/SL),  (SL,H1)).astype(np.float32); b1 = np.zeros(H1,np.float32)
W2 = np.random.uniform(-np.sqrt(2/H1), np.sqrt(2/H1), (H1,H2)).astype(np.float32); b2 = np.zeros(H2,np.float32)
W3 = np.random.uniform(-0.1, 0.1, (H2,1)).astype(np.float32);                       b3 = np.zeros(1, np.float32)

def fwd(Xb):
    Z1  = Xb @ W1 + b1
    AZ1 = np.einsum("ij,bjk->bik", AN, Z1)
    H1r = np.maximum(0, AZ1)
    Z2  = H1r @ W2 + b2
    AZ2 = np.einsum("ij,bjk->bik", AN, Z2)
    H2r = np.maximum(0, AZ2)
    return (H2r @ W3 + b3).squeeze(-1), Z1, AZ1, H1r, Z2, AZ2, H2r

def step(Xb, yb, lr):
    global W1,b1,W2,b2,W3,b3
    yp,Z1,AZ1,H1r,Z2,AZ2,H2r = fwd(Xb)
    d   = 2*(yp-yb)/(len(yb)*NG); d3 = d[:,:,None]
    dW3 = np.einsum("bnh,bno->ho",H2r,d3);  db3 = d3.reshape(-1,1).sum(0)
    dH2 = d3@W3.T;  dAZ2 = dH2*(AZ2>0);    dZ2 = np.einsum("ij,bjk->bik",AN.T,dAZ2)
    dW2 = np.einsum("bnh,bnk->hk",H1r,dZ2); db2 = dZ2.reshape(-1,H2).sum(0)
    dH1 = dZ2@W2.T; dAZ1 = dH1*(AZ1>0);    dZ1 = np.einsum("ij,bjk->bik",AN.T,dAZ1)
    dW1 = np.einsum("bnh,bnk->hk",Xb,dZ1);  db1 = dZ1.reshape(-1,H1).sum(0)
    for W,dW in [(W1,dW1),(W2,dW2),(W3,dW3)]: np.clip(dW,-0.5,0.5,out=dW); W-=lr*dW
    for b,db in [(b1,db1),(b2,db2),(b3,db3)]: np.clip(db,-0.5,0.5,out=db); b-=lr*db.flatten()
    return ((yp-yb)**2).mean()

EP, BG, lr = 60, 256, 0.004
gtl, gtel  = [], []
t0 = time.time()
for ep in range(EP):
    if ep==30: lr=0.002
    if ep==50: lr=0.001
    idx = np.random.permutation(len(Xgtr))
    el, nb = 0, 0
    for s in range(0, len(Xgtr), BG):
        bi = idx[s:s+BG]; el += step(Xgtr[bi], ygtr[bi], lr); nb += 1
    gtl.append(el/nb)
    it = np.random.choice(len(Xgte), 500, replace=False)
    gtel.append(((fwd(Xgte[it])[0] - ygte[it])**2).mean())
    if (ep+1)%15==0:
        print(f"  ep {ep+1:2d}/{EP}  train={gtl[-1]:.5f}  test={gtel[-1]:.5f}  [{time.time()-t0:.0f}s]")

gp = fwd(Xgte)[0]
print(f"\nDone {time.time()-t0:.0f}s | pred [{gp.min():.3f}, {gp.max():.3f}]")
```

```python
# ══════════════════════════════════════════════════════════
# Métriques par capteur + fonctions utilitaires
# ══════════════════════════════════════════════════════════
mae_s  = np.array([mean_absolute_error(ygte[:,i], gp[:,i]) for i in range(NG)])
rmse_s = np.array([np.sqrt(mean_squared_error(ygte[:,i], gp[:,i])) for i in range(NG)])
r2_s   = np.array([1-((ygte[:,i]-gp[:,i])**2).sum()/((ygte[:,i]-ygte[:,i].mean())**2).sum() for i in range(NG)])
mk     = ygte.flatten() > 0.01
mape   = np.mean(np.abs((ygte.flatten()[mk]-gp.flatten()[mk])/ygte.flatten()[mk]))*100

idx_bon     = int(np.argmin(mae_s))
idx_moyen   = int(np.argsort(mae_s)[NG//2])
idx_mauvais = int(np.argmax(mae_s))

def denorm(x):    return x*70 + 5
def cap_color(i): return C["hw"] if gcn_hw[i] else C["ur"] if gcn_ur[i] else C["nm"]
def cap_label(i):
    t = "Autoroute" if gcn_hw[i] else ("Urbain" if gcn_ur[i] else "Nationale")
    return f"S{gcn_s[i]:03d} ({t})"
def err_by_hour(yt, yp, tidx):
    eh = {h:[] for h in range(24)}
    for k in range(len(tidx)): eh[tidx[k].hour].append(abs(yt[k]-yp[k]))
    return {h:np.mean(v) for h,v in eh.items()}

PTS_JOUR    = 288
PTS_SEMAINE = 288*7
PTS_MOIS    = 288*30
test_idx    = pd.date_range("2017-05-01", periods=ygte.shape[0], freq="5min")

print(f"MAE  = {mae_s.mean():.5f}  ({mae_s.mean()*70:.2f} mph)")
print(f"RMSE = {rmse_s.mean():.5f}  ({rmse_s.mean()*70:.2f} mph)")
print(f"MAPE = {mape:.1f}%")
print(f"R²   = {r2_s.mean():.4f}")
print()
print(f"  Meilleur : {cap_label(idx_bon):20s}  MAE={mae_s[idx_bon]*70:.2f}mph  R²={r2_s[idx_bon]:.4f}")
print(f"  Médian   : {cap_label(idx_moyen):20s}  MAE={mae_s[idx_moyen]*70:.2f}mph  R²={r2_s[idx_moyen]:.4f}")
print(f"  Pire     : {cap_label(idx_mauvais):20s}  MAE={mae_s[idx_mauvais]*70:.2f}mph  R²={r2_s[idx_mauvais]:.4f}")
```

```python
# ══════════════════════════════════════════════════════════
# FIGURE A — UNE SEMAINE (1–7 mai 2017)
# (a) Meilleur capteur    (b) Capteur médian
# (c) MAE journalière     (d) Erreur par heure
# (e) Pire capteur        (f) Distribution MAE    (g) Scatter
# ══════════════════════════════════════════════════════════
sl    = slice(0, PTS_SEMAINE)
t_ax  = test_idx[:PTS_SEMAINE]
jours = ["Lun","Mar","Mer","Jeu","Ven","Sam","Dim"]

fig = plt.figure(figsize=(24, 20))
gs  = gridspec.GridSpec(4, 3, hspace=0.48, wspace=0.32,
                        top=0.93, bottom=0.06, left=0.06, right=0.97)
fig.suptitle(
    f"GCN — Analyse sur UNE SEMAINE (1–7 mai 2017)\n"
    f"20 capteurs | MAE moy. = {mae_s.mean()*70:.2f} mph | R² moy. = {r2_s.mean():.4f}",
    fontsize=14, fontweight="bold")

# ── (a) Meilleur capteur ──────────────────────────────────
ax = fig.add_subplot(gs[0, :])
y_t = denorm(ygte[sl, idx_bon]);   y_p = denorm(gp[sl, idx_bon])
ax.plot(range(PTS_SEMAINE), y_t, color=C["true"], lw=1.5, alpha=0.9, label="Vraie vitesse")
ax.plot(range(PTS_SEMAINE), y_p, color=C["gcn"],  lw=1.3, ls="--", alpha=0.8, label="GCN prédiction")
ax.fill_between(range(PTS_SEMAINE), y_t, y_p, alpha=0.18, color=C["err"], label="Erreur")
for d in [5,6]: ax.axvspan(d*PTS_JOUR, min((d+1)*PTS_JOUR,PTS_SEMAINE), alpha=0.07, color=C["weekend"])
for d in range(7):
    ax.axvline(d*PTS_JOUR, color="gray", lw=0.7, ls=":")
    ax.text(d*PTS_JOUR+PTS_JOUR//2, 73, jours[d], ha="center", fontsize=9, color="gray", fontstyle="italic")
ax.set_xticks(np.arange(0,PTS_SEMAINE,PTS_JOUR))
ax.set_xticklabels([f"{j} 00h" for j in jours], fontsize=9)
ax.set_ylabel("Vitesse (mph)", fontsize=11); ax.set_ylim(0,80)
ax.legend(loc="upper right", fontsize=10, framealpha=0.9)
ax.set_title(f"(a) {cap_label(idx_bon)} — Meilleur capteur | MAE={mae_s[idx_bon]*70:.2f}mph  R²={r2_s[idx_bon]:.4f}", fontsize=11)

# ── (b) Capteur médian ────────────────────────────────────
ax = fig.add_subplot(gs[1, :])
y_t = denorm(ygte[sl, idx_moyen]); y_p = denorm(gp[sl, idx_moyen])
ax.plot(range(PTS_SEMAINE), y_t, color=C["true"], lw=1.5, alpha=0.9, label="Vraie vitesse")
ax.plot(range(PTS_SEMAINE), y_p, color=C["gcn"],  lw=1.3, ls="--", alpha=0.8, label="GCN prédiction")
ax.fill_between(range(PTS_SEMAINE), y_t, y_p, alpha=0.18, color=C["err"], label="Erreur")
for d in [5,6]: ax.axvspan(d*PTS_JOUR, min((d+1)*PTS_JOUR,PTS_SEMAINE), alpha=0.07, color=C["weekend"])
for d in range(7):
    ax.axvline(d*PTS_JOUR, color="gray", lw=0.7, ls=":")
    ax.text(d*PTS_JOUR+PTS_JOUR//2, 73, jours[d], ha="center", fontsize=9, color="gray", fontstyle="italic")
ax.set_xticks(np.arange(0,PTS_SEMAINE,PTS_JOUR))
ax.set_xticklabels([f"{j} 00h" for j in jours], fontsize=9)
ax.set_ylabel("Vitesse (mph)", fontsize=11); ax.set_ylim(0,80)
ax.legend(loc="upper right", fontsize=10, framealpha=0.9)
ax.set_title(f"(b) {cap_label(idx_moyen)} — Capteur médian | MAE={mae_s[idx_moyen]*70:.2f}mph  R²={r2_s[idx_moyen]:.4f}", fontsize=11)

# ── (c) MAE journalière ───────────────────────────────────
ax = fig.add_subplot(gs[2, :2])
err_mat = np.zeros((7, NG))
for d in range(7):
    err_mat[d] = np.abs(ygte[d*PTS_JOUR:(d+1)*PTS_JOUR] - gp[d*PTS_JOUR:(d+1)*PTS_JOUR]).mean(axis=0)*70
ax.bar(range(7), err_mat.mean(axis=1),
       color=[C["weekend"] if j>=5 else C["gcn"] for j in range(7)],
       alpha=0.75, edgecolor="white", lw=1.5, zorder=3)
ax.errorbar(range(7), err_mat.mean(axis=1), yerr=err_mat.std(axis=1),
            fmt="none", color="black", capsize=5, lw=1.5, zorder=5)
for j,v in enumerate(err_mat.mean(axis=1)):
    ax.text(j, v+0.05, f"{v:.2f}", ha="center", fontsize=8, fontweight="bold")
ax.set_xticks(range(7)); ax.set_xticklabels(jours, fontsize=10)
ax.set_ylabel("MAE (mph)", fontsize=11); ax.set_ylim(0, err_mat.max()*1.4)
ax.set_title("(c) MAE journalière — moy. 20 capteurs (±1 std inter-capteurs)", fontsize=11)
ax.legend(handles=[Patch(color=C["gcn"],alpha=0.75,label="Jour ouvré"),
                   Patch(color=C["weekend"],alpha=0.75,label="Week-end")], fontsize=9)

# ── (d) Erreur par heure ──────────────────────────────────
ax = fig.add_subplot(gs[2, 2])
errs_h = {h:[] for h in range(24)}
for i in range(NG):
    eh = err_by_hour(ygte[sl,i], gp[sl,i], t_ax)
    for h in range(24): errs_h[h].append(eh[h]*70)
mean_h = [np.mean(errs_h[h]) for h in range(24)]
std_h  = [np.std(errs_h[h])  for h in range(24)]
cols_h = [C["peak"] if (7<=h<=9 or 16<=h<=19) else C["offpeak"] for h in range(24)]
ax.bar(range(24), mean_h, color=cols_h, alpha=0.7, width=0.8)
ax.errorbar(range(24), mean_h, yerr=std_h, fmt="none", color="black", capsize=2, lw=0.8, alpha=0.5)
ax.axvspan(7, 9.5, alpha=0.08, color=C["peak"])
ax.axvspan(16, 19.5, alpha=0.08, color=C["peak"])
ax.set_xlabel("Heure"); ax.set_ylabel("MAE (mph)")
ax.set_title("(d) Erreur par heure de la journée\n(moy. 20 capteurs, 1 semaine)", fontsize=10)
ax.legend(handles=[Patch(color=C["peak"],alpha=0.7,label="Heure de pointe"),
                   Patch(color=C["offpeak"],alpha=0.7,label="Heure creuse")], fontsize=8)

# ── (e) Pire capteur — zoom 3 jours ──────────────────────
ax = fig.add_subplot(gs[3, 0])
n3  = 3*PTS_JOUR
y_t = denorm(ygte[:n3, idx_mauvais]); y_p = denorm(gp[:n3, idx_mauvais])
ax.plot(range(n3), y_t, color=C["true"], lw=1.5, alpha=0.9, label="Vrai")
ax.plot(range(n3), y_p, color=C["err"],  lw=1.3, ls="--", alpha=0.85, label="GCN")
ax.fill_between(range(n3), y_t, y_p, alpha=0.2, color=C["err"])
for d in range(3): ax.axvline(d*PTS_JOUR, color="gray", lw=0.7, ls=":")
ax.set_xticks([0,PTS_JOUR,2*PTS_JOUR,3*PTS_JOUR])
ax.set_xticklabels(["Lun","Mar","Mer","Jeu"], fontsize=8)
ax.set_ylabel("mph"); ax.set_ylim(0,80); ax.legend(fontsize=9)
ax.set_title(f"(e) {cap_label(idx_mauvais)} — Pire capteur\n"
             f"3 premiers jours | MAE={mae_s[idx_mauvais]*70:.2f}mph  R²={r2_s[idx_mauvais]:.4f}", fontsize=9)

# ── (f) Distribution MAE ──────────────────────────────────
ax = fig.add_subplot(gs[3, 1])
ax.hist(mae_s*70, bins=10, color=C["gcn"], alpha=0.75, edgecolor="white", lw=1.5, rwidth=0.85)
ax.axvline(mae_s.mean()*70,    color="red",    lw=2, ls="--", label=f"Moy={mae_s.mean()*70:.2f}mph")
ax.axvline(np.median(mae_s)*70, color="orange", lw=2, ls=":",  label=f"Médiane={np.median(mae_s)*70:.2f}mph")
for i2 in [idx_bon,idx_moyen,idx_mauvais]:
    ax.axvline(mae_s[i2]*70, color=cap_color(i2), lw=1.5, alpha=0.8)
ax.set_xlabel("MAE (mph)"); ax.set_ylabel("Nb capteurs")
ax.set_title("(f) Distribution MAE\n20 capteurs (semaine complète)", fontsize=10)
ax.legend(fontsize=9)

# ── (g) Scatter GCN vs vrai ───────────────────────────────
ax = fig.add_subplot(gs[3, 2])
idx_sc = np.random.choice(PTS_SEMAINE*NG, min(5000,PTS_SEMAINE*NG), replace=False)
y_flat = denorm(ygte[:PTS_SEMAINE].flatten())
p_flat = denorm(gp[:PTS_SEMAINE].flatten())
ax.scatter(y_flat[idx_sc], p_flat[idx_sc], alpha=0.15, s=5, c=C["gcn"])
ax.plot([5,75],[5,75], "r--", lw=2, label="Parfait")
r2_sem = 1-((ygte[:PTS_SEMAINE]-gp[:PTS_SEMAINE])**2).sum()/((ygte[:PTS_SEMAINE]-ygte[:PTS_SEMAINE].mean())**2).sum()
ax.text(0.05,0.92,f"R² = {r2_sem:.4f}",transform=ax.transAxes,fontsize=11,
        bbox=dict(facecolor="white",alpha=0.85,edgecolor=C["gcn"]))
ax.set_xlabel("Vraie vitesse (mph)"); ax.set_ylabel("GCN prédiction (mph)")
ax.set_title("(g) Scatter — GCN vs Vrai\n(1 semaine, 20 capteurs)", fontsize=10)
ax.legend(fontsize=9); ax.set_xlim(0,80); ax.set_ylim(0,80)

plt.savefig("FigGCN_A_Semaine.png", dpi=130, bbox_inches="tight")
plt.show()
print("Sauvegardé : FigGCN_A_Semaine.png")
```
